```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fold_down = "../roberto/DOWN/"
fold_left = "../roberto/LEFT/"
fold_right = "../roberto/RIGHT/"
fold_up = "../roberto/UP/"
```

```python
down = pd.read_csv(fold_down + 'down1.csv', index_col=0)

left = pd.read_csv(fold_left + 'left1.csv', index_col=0)

right = pd.read_csv(fold_right + 'right1.csv', index_col=0)

up = pd.read_csv(fold_up + 'up1.csv', index_col=0)

down.head()
```

```python
fig, axes = plt.subplots(3, 1, figsize=(14, 18), sharex=True)

# Primo grafico: GyroX, GyroY, GyroZ
axes[0].plot(down.index, down['GyroX'], label='GyroX', color='blue')
axes[0].plot(down.index, down['GyroY'], label='GyroY', color='green')
axes[0].plot(down.index, down['GyroZ'], label='GyroZ', color='red')
axes[0].set_title('Gyro - down')
axes[0].set_ylabel('Valore')
axes[0].legend()
axes[0].grid(True)

# Secondo grafico: AccX, AccY, AccZ
axes[1].plot(down.index, down['AccX'], label='AccX', color='purple')
axes[1].plot(down.index, down['AccY'], label='AccY', color='orange')
axes[1].plot(down.index, down['AccZ'], label='AccZ', color='brown')
axes[1].set_title('Acc - down')
axes[1].set_xlabel('N° record')
axes[1].set_ylabel('Valore')
axes[1].legend()
axes[1].grid(True)

# Terzo grafico: Gyro + Acc
axes[2].plot(down.index, down['GyroX'], label='GyroX', color='blue')
axes[2].plot(down.index, down['GyroY'], label='GyroY', color='green')
axes[2].plot(down.index, down['GyroZ'], label='GyroZ', color='red')
axes[2].plot(down.index, down['AccX'], label='AccX', color='purple')
axes[2].plot(down.index, down['AccY'], label='AccY', color='orange')
axes[2].plot(down.index, down['AccZ'], label='AccZ', color='brown')
axes[2].set_title('Gyro + Acc - down')
axes[2].set_ylabel('Valore')
axes[2].legend()
axes[2].grid(True)

for ax in axes:
    ax.set_xticks(np.arange(0, len(down), step=25))
    ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.savefig("confronto_distr/down1")
plt.show()
```
