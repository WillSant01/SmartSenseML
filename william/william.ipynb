```python
# CAMBIO MEDIA CON MINIMO E MASSIMO

def custom_aggregation(series):
    positive_count = (series > 0).sum()
    negative_count = (series < 0).sum()
    
    if positive_count > negative_count:
        return series.max()
    else:
        return series.min()

up_down_minmax = df_up_down.resample('1S').apply(custom_aggregation)
right_left_minmax = df_right_left.resample('1S').apply(custom_aggregation)
box_minmax = box.resample('1S').apply(custom_aggregation)
```

```python
plt.figure(figsize=(14, 7))
plt.plot(up_down_minmax.index, up_down_minmax['GyroX'], label='GyroX', color='blue')
plt.plot(up_down_minmax.index, up_down_minmax['GyroY'], label='GyroY', color='green')
plt.plot(up_down_minmax.index, up_down_minmax['GyroZ'], label='GyroZ', color='red')

plt.xlabel('Timestamp')
plt.ylabel('Valore')
plt.title('GyroX, GyroY, GyroZ 1 S UP DOWN MINMAX')
plt.legend()
plt.grid(True)

plt.show()
```
